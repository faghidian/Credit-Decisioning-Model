```python
import pandas as pd
import numpy as np

# Load the dataset A
dfA = pd.read_csv("D:/file A.csv", header=0)
display(dfA.head())

# Inspect the dataset structure
display(dfA.info())
```

```python
# Convert SPENDING to numeric (handling any potential errors with coercion)
dfA['SPENDING'] = pd.to_numeric(dfA['SPENDING'], errors='coerce').round()

# Convert APPLICATION_DATE to datetime format, handling errors
dfA['APPLICATION_DATE'] = pd.to_datetime(dfA['APPLICATION_DATE'], errors='coerce')

#  Extract the first part of the 'ID' by splitting the string on '--' and Convert 'ID' to int64
dfA['ID'] = dfA['ID'].str.split('--').str[0].astype('int64')

display(dfA.head())
display(dfA.info())
```

```python
# Function to count null values and calculate the percentage of nulls
def NullCheck(data):
    d = dict()
    for col in data.columns:
        d[col] = [data[col].isnull().sum(), round(data[col].isnull().mean()*100,3)]
    ddd = pd.DataFrame(d).T
    ddd = ddd.rename(columns={0: "NullSum", 1: "NullMean%"})
    return ddd.sort_values(by="NullSum", ascending=False)

NullCheck(dfA)
```

```python
# Count duplicate IDs
print(f"Number of duplicate IDs:")
dfA.duplicated(subset=['ID']).sum()
```

```python
# Load the dataset B
dfB = pd.read_csv("D:/file B.csv", header=0)
display(dfB.head())

# Inspect the dataset structure
display(dfB.info())
```

```python
NullCheck(dfB)
```

```python
# Count duplicate IDs
print(f"Number of duplicate IDs:")
dfB.duplicated(subset=['ID']).sum()
```

```python
# Drop duplicates based on 'ID', keeping the first occurrence
dfB = dfB.drop_duplicates(subset=['ID'], keep='first')

# Count any remaining fully duplicated rows
print(f"Number of duplicate IDs:")
dfB.duplicated().sum()
```

```python
# Round AGE to the nearest whole number
dfB['AGE'] = dfB['AGE'].round().astype('Int64')

# drop the birth_year column (as AGE is sufficient)
dfB.drop(columns=['birth_year'], inplace=True)
```

```python
# Check correlation between MONTHLY_INCOME and ANNUAL_INCOME
correlation = dfB[['MONTHLY_INCOME', 'ANNUAL_INCOME']].corr()
display("Correlation between MONTHLY_INCOME and ANNUAL_INCOME:\n", correlation)
```

```python
# Drop 'MONTHLY_INCOME' column as it is highly corollated with 'ANNUAL_INCOME' 
dfB.drop(columns=['MONTHLY_INCOME'], inplace=True)
```

The columns related to "lag" values (e.g., `MAJORDRG_lag1` to `MAJORDRG_lag12` and `MINORDRG_lag1` to `MINORDRG_lag12`) are typically used in time-series analysis or feature engineering to capture trends over time. However, for this credit decisioning model, these lag variables may not be necessary and may introduce multicollinearity or redundancy when building the model.

```python
# Identify columns related to 'lag' in their name
lag_columns = [col for col in dfB.columns if 'lag' in col]

# Drop the lag columns from dfB
dfB.drop(columns=lag_columns, inplace=True)

# Display the updated dfB to verify the lag columns are dropped
display(dfB.head())
```

```python
# Merge dfA and dfB on the 'ID' column
# Ensure 'ID' column is the same type in both DataFrames before merging
dfA['ID'] = dfA['ID'].astype('Int64')
dfB['ID'] = dfB['ID'].astype('Int64')

# Merge dfA and dfB on the 'ID' column
df = pd.merge(dfA, dfB, on='ID', how='inner')

# Display the first few rows of the merged dataframe to verify the result
display(df.head())

# Inspect the dataset structure
display(df.info())
```

```python
# Count duplicate IDs
print(f"Number of duplicate IDs:")
df.duplicated().sum()
```

```python
NullCheck(df)
```

```python
from sklearn.impute import KNNImputer

def knn_impute(data, collst, n_neighbors=5):
    imputer = KNNImputer(n_neighbors=n_neighbors)
    
    # Iterate through columns and apply KNN imputation
    for col in collst:
        if data[col].dtype in [np.float64, np.int64]:
            data_col = data[[col]]
            print(f"Imputing missing values in column '{col}' using KNNImputer.")
            data[[col]] = imputer.fit_transform(data_col)
        else:
            print(f"Column '{col}' is not numerical, skipping imputation.")
    return data

# Identify columns with missing values
dfnch = NullCheck(df)
lst = list(dfnch[dfnch['NullMean%'] != 0].index)

# Apply KNN imputation to those columns
df = knn_impute(df, lst)
```

```python
NullCheck(df)
```

```python
df.head()
```

```python
# Ensure the APPLICATION_DATE is in datetime format
df['APPLICATION_DATE'] = pd.to_datetime(df['APPLICATION_DATE'], format='%Y-%m-%d')

# Calculate the month_since_application
df['MTH_SINCE_APP'] = ((pd.to_datetime('today') - df['APPLICATION_DATE']).dt.days // 30).round()

# Calculate Age of the Applicant at the Time of Application
# Subtract the years between today and APPLICATION_DATE from the current age
df['AGE_AT_APP'] = df['AGE'] - ((pd.to_datetime('today') - df['APPLICATION_DATE']).dt.days // 365)

# Drop the APPLICATION_DATE column as it's no longer needed
df.drop(columns=['APPLICATION_DATE'], inplace=True)

# Display the updated dataframe
display(df[['MTH_SINCE_APP', 'AGE_AT_APP']].head())
```

```python
# List of columns to convert to integers
columns_to_convert = ['SPENDING', 'ANNUAL_INCOME']

# Convert each column in the list to 'int64'
for column in columns_to_convert:
    df[column] = df[column].astype('int64')
```
